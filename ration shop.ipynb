```python
import random

# Initialize user credits
monthly_credits = 100
current_credits = monthly_credits

# Sample goods data with prices in INR and credit points
goods_data = {
    "rice": {"quantity": 20, "price": 35, "credits": 10},
    "dhall": {"quantity": 15, "price": 30, "credits": 5},
    "sugar": {"quantity": 25, "price": 15, "credits": 15},
    "oil": {"quantity": 10, "price": 20, "credits": 20}
}

# Initialize user's cart
cart = {}

def list_available_items():
    available_items = "\n".join([f"{item}: {data['quantity']} available, Price: ₹{data['price']}, Credits: {data['credits']}" for item, data in goods_data.items()])
    return f"Available items:\n{available_items}"

def allocate_goods(item, quantity, current_credits):
    available_quantity = goods_data.get(item, {}).get('quantity', 0)
    item_credits = goods_data.get(item, {}).get('credits', 0)
    
    if available_quantity > 0:  # Check if the item is available
        if available_quantity >= quantity:
            if item_credits * quantity <= current_credits:
                goods_data[item]['quantity'] -= quantity
                current_credits -= item_credits * quantity
                cart[item] = cart.get(item, 0) + quantity  # Update cart
                token = random.randint(1000, 9999)
                return f"Allocated {quantity} {item}(s) for {item_credits * quantity} credits. Your token: {token}"
            else:
                return f"You don't have enough credits for {quantity} {item}(s)."
        else:
            return f"Sorry, only {available_quantity} {item}(s) available."
    else:
        return f"Sorry, {item} is not available."

def view_credits(current_credits):
    return f"Available credits: {current_credits}"

# Show menu at the beginning
print("Chatbot:")
print("1. List available items")
print("2. Allocate items")
print("3. View cart")
print("4. View available credits")
print("5. Exit")

# Simple chatbot interaction
while True:
    choice = input("Choose an option: ")

    if choice == '1':
        print(list_available_items())
    elif choice == '2':
        item = input("Enter the item you want to allocate: ")
        quantity = int(input(f"Enter the quantity of {item} you want to allocate: "))
        response = allocate_goods(item, quantity, current_credits)
        print(response)
        current_credits -= quantity * goods_data.get(item, {}).get('credits', 0)
    elif choice == '3':
        if cart:
            total_credits_used = 0
            print("Your cart:")
            for item, quantity in cart.items():
                item_credits = goods_data[item]['credits']
                item_total_credits = item_credits * quantity
                total_credits_used += item_total_credits
                print(f"{item}: {quantity} @ {item_credits} credits each = {item_total_credits} credits")
            print(f"Total credits used: {total_credits_used}")
            print(f"Remaining credits: {current_credits}")
        else:
            print("Your cart is empty.")
    elif choice == '4':
        print(view_credits(current_credits))
    elif choice == '5':
        print("Exiting...")
        break
    else:
        print("Invalid choice. Please choose a valid option.")
```
